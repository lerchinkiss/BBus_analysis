```python
import pandas as pd
```

```python
data = pd.read_csv('cars.csv',  delimiter=',')
```

```python
###Выведем первые 15 значений

data.head(15)
```

```python
###Выведем последние 15 значений

data.tail(15)
```

```python
###Проверка пропущенных значений

data.isnull().sum()
```

```python
###Проверка все ли пропущены значения в колонке fuel_type

data['fuel_type']
```

```python
###Проверка все ли пропущены значения в колонке average_text

data['average_text']
```

```python
###Заменим пропущенные значения в столбце - fuel_type на Gas

data['fuel_type'] = data['fuel_type'].fillna('Gas')
data['fuel_type'].unique()
```

```python
###Заменим пропущенные значения в столбце - average_text на No answer

data['average_text'] = data['average_text'].fillna('No answer')
data['average_text'].unique()
```

```python
###Проверка пропущенных значений после обработки 

data.isnull().sum()
```

```python
###Наиболее часто бронируемые машины

most_booked_cars = data['product_name'].value_counts().head(10)
print(most_booked_cars)
```

```python
###Максимальная и минимальная цена аренды

max_price = data['price'].max()
min_price = data['price'].min()

print(f"Максимальная цена: {max_price}")
print(f"Минимальная цена: {min_price}")
```

```python
###Средняя цена аренды по поставщикам

average_price_by_supplier = data.groupby('supplier_name')['price'].mean().round(2).sort_values(ascending=False).head(10)

print(average_price_by_supplier)
```

```python
###Средний рейтинг поставщиков

average_rating_by_supplier = data.groupby('supplier_name')['average'].mean().round(2).sort_values(ascending=False).head(10)

print(average_rating_by_supplier)
```

```python
# Описательная статистика для цен
print(data['price'].describe())

# Построение гистограммы
import matplotlib.pyplot as plt
data['price'].hist(bins=20, edgecolor='black')
plt.title('Распределение цен')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.show()
```

```python
# Популярные города
top_cities = data['city'].value_counts().head(10)
print("Топ-10 городов:")
print(top_cities)

# Популярные страны
top_countries = data['country'].value_counts().head(10)
print("Топ-10 стран:")
print(top_countries)
```

```python
# Средняя цена по типу топлива
average_price_by_fuel = data.groupby('fuel_type')['price'].mean().round(2)
print(average_price_by_fuel)

# Средняя цена по количеству мест
average_price_by_seats = data.groupby('seats')['price'].mean().round(2)
print(average_price_by_seats)
```
